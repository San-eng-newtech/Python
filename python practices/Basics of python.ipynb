```python
class solution:
    def fizbuzz(self, n:int) -> list[str]:
        ans = []
        for i in range(1,n+1):
            if i%3==0 and i%5==0:
                ans.append("FizzBuzz")
            elif i%3==0:
                ans.append("Fizz")
            elif i%5==0:
                ans.append("Buzz")
            else:
                ans.append(str(i))
        return ans

obj = solution()
result = obj.fizbuzz(15)

print(result)
```

```python
#given two non-negative integers low and high.return the count of odd number between  low and high count odd numbers
# in an interval range  

class solution:
    def countOdds(self, low:int, high:int) -> int:
        return (high + 1)//2 - (low // 2)
        

obj = solution()
reult = obj.countOdds(5,19)
print(reult)
```

```python
#how many numbers are smaller then the current numbers
nums = [8, 1, 2, 2, 3]

def smallerNumbersThanCurrent(nums: list[int]) -> list[int]:
    ans = []
    for i in nums:
        c = 0
        for j in nums:
            if j < i:
                c += 1
        ans.append(c)
    return ans

s = smallerNumbersThanCurrent(nums)
print(s)
```

```python
## count the digits that divide a numbers 

def countDigits(num:int) -> int:
    temp = num
    ans = 0
    while temp>0:
        r = temp%10
        if num%r==0:
            ans+=1
        temp//=10
    return ans

s = countDigits(45435)
print(s)
```

```python
#PALINDROME NUMBER 
def isPalindrome(x: int) -> bool:
    temp = x
    rev = 0

    while temp > 0:
        r = temp % 10
        temp //= 10
        rev = rev * 10 + r

    return x == rev

print(isPalindrome(121))   # True
print(isPalindrome(123))   # False
```

```python
#Subtracrt the product and sum of digits of an integer

def subtractProductAndSum(n:int) -> int:
    temp = n
    sum_ = 0
    product = 1
    while temp>0:
        r = temp%10
        temp//=10
        sum_+=r
        product*=r
    return product-sum_


print(subtractProductAndSum(10))
```

```python
#kids with greatest number of candies 

def kidswithcandies(candies: list[int], extracandies: int) -> list[bool]:
    maxcandies = max(candies)
    ans = []

    for i in candies:
        if (i + extracandies) >= maxcandies:
            ans.append(True)
        else:
            ans.append(False)
    return ans

print(kidswithcandies([2, 3, 5, 1, 3], 3))
```
