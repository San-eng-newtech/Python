```python
##running sum of 1d arraay 

def runningSum(nums:list[int])->list[int]:
    n = len(nums)

    ans=[]
    ans.append(nums[0]) 

    for i in range(1,n):
        x = ans[i-1] + nums[i]
        ans.append(x)

    return ans 

print(runningSum([1,2,3,4,5]))
```

```python
 # Remove duplicates from sorted array
def removeDuplicates(nums: list[int]) -> int:
    n = len(nums)

    start = 0
    for i in range(1, n):
        if nums[i] != nums[start]:
            start += 1
            nums[start] = nums[i]

    return start + 1


print(removeDuplicates([1,1,2,2,3,3,4,4,5]))
```

```python
## Sort Array by Parity 

def sortarrbyParity(nums:list[int])->list[int]:
    n = len(nums)
    start = 0
    for i in range(n):
        if nums[i]%2==0:
            temp=nums[i]
            nums[i] = nums[start]
            nums[start] = temp 
            start+=1
    return nums 

print(sortarrbyParity([1,2,3,45,6,6]))
```

```python
##Maximum Subarray 

def maxSubArray(nums: list[int]) -> int:
    curr_sum = 0
    max_sum = nums[0]

    for i in range(len(nums)):
        curr_sum += nums[i]

        if curr_sum > max_sum:
            max_sum = curr_sum

        if curr_sum < 0:
            curr_sum = 0

    return max_sum


print(maxSubArray([1,2,3,4,5,6,7,8]))
```

```python
#Best time to buy and sell stock 
def maxProfit(prices:list[int])->int:
    min_price = prices[0]
    profit = 0

    for i in range(1,len(prices)):
        curr_profit = prices[i] - min_price
        if curr_profit>profit:
            profit  = curr_profit
        min_price = min(min_price,prices[i])
    return profit 

print(maxProfit((20,30,40,405,10)))
```

```python
def maxProfit(prices:list[int])->int:
    profit = 0
    n = len(prices)

    for i in range(n-1):
        if prices[i+1]>prices[i]:
            profit += (prices[i+1]-prices[i])

    return profit 

print(maxProfit((20,30,304,50)))
    
```

```python
## Richest customer wealth 

def maxiWealth(accounts: list[list[int]]) -> int:
    ans = 0
    for account in accounts:
        ans = max(ans, sum(account))
    return ans


print(maxiWealth([[20, 30], [400, 3000]]))
```

```python
#Spiral Matrix 

def spiralOrder(matrix: list[list[int]]) -> list[int]:
    n = len(matrix)
    m = len(matrix[0])

    total = n * m
    ans = []
    c = 0

    rowstart = 0
    rowend = n - 1
    colstart = 0
    colend = m - 1

    while c < total:

        # top row
        for i in range(colstart, colend + 1):
            ans.append(matrix[rowstart][i])
            c += 1
        rowstart += 1
        if c == total:
            break

        # right column
        for i in range(rowstart, rowend + 1):
            ans.append(matrix[i][colend])
            c += 1
        colend -= 1
        if c == total:
            break

        # bottom row
        for i in range(colend, colstart - 1, -1):
            ans.append(matrix[rowend][i])
            c += 1
        rowend -= 1
        if c == total:
            break

        # left column
        for i in range(rowend, rowstart - 1, -1):
            ans.append(matrix[i][colstart])
            c += 1
        colstart += 1

    return ans


print(spiralOrder([[1,2,3],[4,5,6],[7,8,9]]))
            
```

```python
# Defanging an IP Address

def defang(address: str) -> str:
    ans = ""
    for i in address:
        if i != ".":
            ans += i
        else:
            ans += "[.]"
    return ans


print(defang('1.2.3.3.4'))
```

```python
# Valid Palindrome 

def isAlphanumeric(s):
    x = ord(s)
    if 97 <= x <= 122 or 65 <= x <= 90 or 48 <= x <= 57:
        return True
    return False


def isPalindrome(s: str) -> bool:
    s = s.lower()
    i = 0
    j = len(s) - 1

    while i < j:
        if not isAlphanumeric(s[i]):
            i += 1
        elif not isAlphanumeric(s[j]):
            j -= 1
        elif s[i] == s[j]:
            i += 1
            j -= 1
        else:
            return False

    return True

print(isPalindrome("A man, a plan, a canal: Panama"))
```

```python
## Reverse String 

def ReverseString(s: list[str]) -> None:
    i = 0
    j = len(s) - 1

    while i < j:
        temp = s[i]
        s[i] = s[j]
        s[j] = temp

        i += 1
        j -= 1

s = ["h", "e", "l", "l", "o"]

ReverseString(s)

print(s)
    
    
```
