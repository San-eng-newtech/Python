```python
num = 5
factorial = 1

for i in range(1, num):
    factorial *= i

print("Factorial of", num, "is", factorial)
```

```python
def fib(n:int) -> int:
    #base case
    if n==0 or n==1:
        return n
        #recursive case 
    return fib(n-1) + fib(n-2)

print(fib(7))
```

```python
def trib(n: int) -> int:
    # base cases
    if n == 0:
        return 0
    elif n == 1 or n == 2:
        return 1
    
    # recursive case
    return trib(n-1) + trib(n-2) + trib(n-3)

print(trib(7))
```

```python
##GIVEN an integer n, return ture if its a power of two.otherwide,return false

def isPowerofTwo(n:int)-> bool:
    if (n<0):
        return False
    while n%2==0:
        n//2

    if n==1:
        return True 
    else:
        return False

print(isPowerofTwo(29))
print(isPowerofTwo(56))
```

```python
## GCD OF TWO  NUMBERS- Euclidean algorithm 

def gcd(a,b):
    if b==0:
        return a 
    return gcd(b,a%b)

def lcm(a,b):
    return a*b//gcd(a,b)


print(gcd(15,30)) 
print(lcm(10,20))
```

```python
def findPow(x, n):
    if n == 0:
        return 1
    
    a = findPow(x, n // 2)
    
    if n % 2 == 0:
        return a * a
    else:
        return a * a * x

def Mpow(x: float, n: int) -> float:
    if n >= 0:
        return findPow(x, n)
    else:
        if x == 0:
            raise ValueError("0 cannot be raised to a negative power")
        return 1 / findPow(x, -n)
    
```

```python
print(Mpow(2, 10))   
print(Mpow(2, -2))   
print(Mpow(5, 3))    
print(Mpow(3, 0))    
```
